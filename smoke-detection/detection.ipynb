```python
import numpy as np
import cv2
from vidgear.gears import CamGear
import time

sources = {"asl": "https://www.youtube.com/watch?v=mglAxTiW0wE",
           "asl2": "https://www.youtube.com/watch?v=TKLS8aDAyOU",
           "george": "https://www.youtube.com/embed/DtWQiDlnPuU"}

stream = CamGear(source=sources["asl2"], y_tube =True).start()

frame = stream.read()
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
last_frame = gray

height, width = gray.shape[:2]
cv2.namedWindow('jpg', cv2.WINDOW_NORMAL)
cv2.resizeWindow('jpg', 2*height, 2*width)

while True:

    if frame is None:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    diff = cv2.absdiff(gray, last_frame)
    cv2.imshow('jpg',diff)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    time.sleep(0.01)
    frame = stream.read()

cv2.destroyAllWindows()
stream.stop()
```

```python
# uso de uma máscara de todos os arquivos
import numpy as np
import cv2
import time
import os

dir = '/home/palmdev/projects/dsp/dsp-course/smoke-detection/dataset/'
files = os.listdir(dir)
avi_files = []

for names in files:
    if names.endswith(".avi"):
        avi_files.append(names)
        cap = cv2.VideoCapture(dir + names)
        ret, frame = cap.read()
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        last_frame = gray

        height, width = gray.shape[:2]
        cv2.namedWindow('jpg', cv2.WINDOW_NORMAL)
        cv2.resizeWindow('jpg', 2*height, 2*width)

        while (cap.isOpened()):

            if frame is None:
                break

            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

            diff = cv2.absdiff(gray, last_frame)


            th, mask = cv2.threshold(diff, 80, 255, cv2.THRESH_BINARY)
            
            cv2.imshow('jpg',mask)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break


            #time.sleep(0.05)
            ret, frame = cap.read()

        cap.release()
        cv2.destroyAllWindows()
```

```python
# Uso de Erosão

cap = cv2.VideoCapture(dir + avi_files[1])

ret, frame = cap.read()
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
last_frame = gray

height, width = gray.shape[:2]
cv2.namedWindow('jpg', cv2.WINDOW_NORMAL)
cv2.resizeWindow('jpg', 4*height, 4*width)

while (cap.isOpened()):

    if frame is None:
        break


    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    diff = cv2.absdiff(gray, last_frame)


    th, threshold = cv2.threshold(diff, 80, 255, cv2.THRESH_BINARY)
    
    kernel = np.ones((5,5),np.uint8)
    dilation = cv2.dilate(threshold,kernel,iterations = 1)

    cv2.imshow('jpg',dilation)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


    time.sleep(0.05)
    ret, frame = cap.read()

cap.release()
cv2.destroyAllWindows()
```

```python
# Uso de Dilatação

cap = cv2.VideoCapture(dir + avi_files[1])

ret, frame = cap.read()
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
last_frame = gray

height, width = gray.shape[:2]
cv2.namedWindow('jpg', cv2.WINDOW_NORMAL)
cv2.resizeWindow('jpg', 4*height, 4*width)

while (cap.isOpened()):

    if frame is None:
        break


    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    diff = cv2.absdiff(gray, last_frame)


    th, threshold = cv2.threshold(diff, 80, 255, cv2.THRESH_BINARY)
    
    kernel = np.ones((5,5),np.uint8)
    erosion = cv2.erode(threshold,kernel,iterations = 1)

    cv2.imshow('jpg',erosion)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


    time.sleep(0.05)
    ret, frame = cap.read()

cap.release()
cv2.destroyAllWindows()
```
