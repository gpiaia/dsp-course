```python
import numpy as np
import cv2
from vidgear.gears import CamGear
import time

sources = {"asl": "https://www.youtube.com/watch?v=mglAxTiW0wE",
           "asl2": "https://www.youtube.com/watch?v=TKLS8aDAyOU",
           "george": "https://www.youtube.com/embed/DtWQiDlnPuU"}

stream = CamGear(source=sources["asl2"], y_tube =True).start()

frame = stream.read()
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
last_frame = gray

height, width = gray.shape[:2]
cv2.namedWindow('screen', cv2.WINDOW_NORMAL)
cv2.resizeWindow('screen', 2*height, 2*width)

while True:

    if frame is None:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    diff = cv2.absdiff(gray, last_frame)
    cv2.imshow('screen',diff)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    time.sleep(0.01)
    frame = stream.read()

cv2.destroyAllWindows()
stream.stop()
```

```python
import numpy as np
import cv2
import time
import os

dir = '/home/palmdev/projects/dsp/dsp-course/smoke-detection/dataset/'
files = os.listdir(dir)
avi_files = []

for names in files:
    if names.endswith(".avi"):
        avi_files.append(names)

cap = cv2.VideoCapture(dir + avi_files[9])
ret, frame = cap.read()
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
last_frame = gray

height, width = gray.shape[:2]
cv2.namedWindow('screen', cv2.WINDOW_NORMAL)
cv2.resizeWindow('screen', 4*height, 4*width)

while (cap.isOpened()):

    if frame is None:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    diff = cv2.absdiff(gray, last_frame)

    th, threshold = cv2.threshold(diff, 80, 255, cv2.THRESH_BINARY)
    
    kernel = np.ones((5,5),np.uint8)
    
    dilation= cv2.dilate(threshold, kernel, iterations = 1)
    erosion = cv2.erode(dilation, kernel, iterations = 1)

    mask = np.zeros(dilation.shape,np.uint8)

    contours, hier = cv2.findContours(erosion,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)

    for cnt in contours:
        if 0<cv2.contourArea(cnt)<5000:
            cv2.drawContours(frame,[cnt],0,(0,255,0),2)
            cv2.drawContours(mask,[cnt],0,255,-1)
    
    cv2.imshow('screen', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    time.sleep(0.05)
    ret, frame = cap.read()

cap.release()
cv2.destroyAllWindows()
```

## Transformada de Fourier e Standard Deviation

```python
def histogram(I):
    dft = cv2.dft(np.float32(I), flags=cv2.DFT_COMPLEX_OUTPUT)
    dft_abs = np.abs(dft)
    
    histogram = np.histogram(dft_abs)

    return histogram 

def feature2(blob):
    I = np.matrix(blob)
    I = I.max(1)

    I_histogram = histogram(I)

    M = np.ndarray.max(I_histogram[0])
    return M

def feature1(frame):
    std = np.std(frame, axis = 0)
    return np.ndarray.max(std)
```
