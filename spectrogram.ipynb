```python
from scipy.signal import chirp, spectrogram
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
Fs = 2e3;
Np = 4e3;

incremento=1/Fs
t = np.linspace(0, incremento*(Np-1), Np)

f1 = 1
f0 = 1e2

t1 = 0.75

w = chirp(t, f0, f1, t1, method='quadratic')

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,8))
plt.plot(t, w)
plt.title("Linear Chirp, f(0)=6, f(10)=1")
plt.xlabel('t (sec)')
```

```python
ff, tt, Sxx = spectrogram(w, fs=Fs, noverlap=500, nperseg=512, nfft=2048)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,8))
plt.pcolormesh(tt, ff, Sxx)
plt.title('Linear Chirp, f(0)=1500, f(10)=250')
plt.xlabel('t (sec)')
plt.ylabel('Frequency (Hz)')
plt.grid()
```
